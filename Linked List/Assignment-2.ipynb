```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    # Function to display the linked list
    def display(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            current = self.head
            while current:
                print(current.data, end=" -> " if current.next else "\n")
                current = current.next
    
    # Function to insert at the beginning
    def insert_at_start(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    
    # Function to insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
    
    # Function to delete from the start
    def delete_from_start(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        else:
            print(f"Deleting {self.head.data} from the start")
            self.head = self.head.next
    
    # Function to delete from the end
    def delete_from_end(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        elif self.head.next is None:
            print(f"Deleting {self.head.data} from the end")
            self.head = None
        else:
            current = self.head
            while current.next.next:
                current = current.next
            print(f"Deleting {current.next.data} from the end")
            current.next = None

# Create a linked list and initialize it with the given values
linked_list = LinkedList()

# Inserting the initial values into the linked list
initial_values = [5, 10, 15, 20, 25]
for value in initial_values:
    linked_list.insert_at_end(value)

# Display the initial linked list
print("Initial Linked List:")
linked_list.display()

# Task 1: Insert values 30 and 35 at the start
print("\nInserting 30 and 35 at the start:")
linked_list.insert_at_start(30)
linked_list.insert_at_start(35)
linked_list.display()

# Task 2: Insert values 40 and 45 at the end
print("\nInserting 40 and 45 at the end:")
linked_list.insert_at_end(40)
linked_list.insert_at_end(45)
linked_list.display()

# Task 3: Delete a value from the start
print("\nDeleting value from the start:")
linked_list.delete_from_start()
linked_list.display()

# Task 4: Delete a value from the end
print("\nDeleting value from the end:")
linked_list.delete_from_end()
linked_list.display()
```

```python
class Stack:
    def __init__(self):
        self.stack = [18, 3, 12, 7, 25, 1, 6, 13]  # Initial stack

    def push(self, item):
        self.stack.append(item)  # Add item to the top

    def pop(self):
        return self.stack.pop() if not self.is_empty() else "Stack is empty"

    def peek(self):
        return self.stack[-1] if not self.is_empty() else "Stack is empty"

    def is_empty(self):
        return len(self.stack) == 0

    def quick_sort(self):
        self.stack = self._quick_sort_recursive(self.stack)

    def _quick_sort_recursive(self, stack):
        if len(stack) <= 1:
            return stack
        else:
            pivot = stack.pop()  # Select the last element as pivot
            greater = [x for x in stack if x > pivot]  # Elements greater than pivot
            lesser = [x for x in stack if x <= pivot]  # Elements lesser than pivot
            # Recursively sort both parts and combine with pivot in the middle
            return self._quick_sort_recursive(greater) + [pivot] + self._quick_sort_recursive(lesser)

    def display(self):
        print(self.stack)

# Task 1: Sorting the stack
stack = Stack()
stack.quick_sort()  # Sort the stack in descending order
stack.display()  # Output: [25, 18, 13, 12, 7, 6, 3, 1]
```

### Sorting a Stack using Quick Sort (Descending Order)

## customer service

```python
class Customer:
    def __init__(self, customer_id, issue_description):
        self.customer_id = customer_id
        self.issue_description = issue_description

    def __repr__(self):
        return f"Customer ID: {self.customer_id}, Issue: {self.issue_description}"

class CustomerQueue:
    def __init__(self):
        self.queue = []

    # Enqueue a customer (add to the queue)
    def enqueue(self, customer_id, issue_description):
        new_customer = Customer(customer_id, issue_description)
        self.queue.append(new_customer)

    # Dequeue a customer (remove the first customer in line)
    def dequeue(self):
        if self.is_empty():
            return "No customers in queue."
        return self.queue.pop(0)

    # Display the current queue of customers
    def display_queue(self):
        if self.is_empty():
            print("The queue is empty.")
        else:
            print("Current Queue of Customers:")
            for customer in self.queue:
                print(customer)

    # Check if the queue is empty
    def is_empty(self):
        return len(self.queue) == 0

    # Insertion sort based on customer_id (ascending order)
    def insertion_sort(self):
        for i in range(1, len(self.queue)):
            key_customer = self.queue[i]
            j = i - 1
            while j >= 0 and self.queue[j].customer_id > key_customer.customer_id:
                self.queue[j + 1] = self.queue[j]
                j -= 1
            self.queue[j + 1] = key_customer

    # Periodically sort the queue based on customer_id
    def sort_queue(self):
        print("\nSorting the queue based on Customer IDs...")
        self.insertion_sort()
        self.display_queue()

# Create an instance of the CustomerQueue class
customer_service_queue = CustomerQueue()

# Add customers to the queue
customer_service_queue.enqueue(102, "Internet connection issue")
customer_service_queue.enqueue(101, "Billing problem")
customer_service_queue.enqueue(104, "Unable to login")
customer_service_queue.enqueue(103, "Slow speed issue")

# Display the initial queue
print("Initial Queue:")
customer_service_queue.display_queue()

# Sort the queue based on customer ID
customer_service_queue.sort_queue()

# Dequeue a customer and display the queue
print("\nAfter serving one customer:")
customer_service_queue.dequeue()
customer_service_queue.display_queue()

# Add a new customer and display the queue
print("\nAdding a new customer to the queue:")
customer_service_queue.enqueue(100, "New account setup")
customer_service_queue.display_queue()

# Sort the queue again after adding the new customer
customer_service_queue.sort_queue()
```
