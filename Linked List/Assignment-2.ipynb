```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    # Function to display the linked list
    def display(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            current = self.head
            while current:
                print(current.data, end=" -> " if current.next else "\n")
                current = current.next
    
    # Function to insert at the beginning
    def insert_at_start(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    
    # Function to insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
    
    # Function to delete from the start
    def delete_from_start(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        else:
            print(f"Deleting {self.head.data} from the start")
            self.head = self.head.next
    
    # Function to delete from the end
    def delete_from_end(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        elif self.head.next is None:
            print(f"Deleting {self.head.data} from the end")
            self.head = None
        else:
            current = self.head
            while current.next.next:
                current = current.next
            print(f"Deleting {current.next.data} from the end")
            current.next = None

# Create a linked list and initialize it with the given values
linked_list = LinkedList()

# Inserting the initial values into the linked list
initial_values = [5, 10, 15, 20, 25]
for value in initial_values:
    linked_list.insert_at_end(value)

# Display the initial linked list
print("Initial Linked List:")
linked_list.display()

# Task 1: Insert values 30 and 35 at the start
print("\nInserting 30 and 35 at the start:")
linked_list.insert_at_start(30)
linked_list.insert_at_start(35)
linked_list.display()

# Task 2: Insert values 40 and 45 at the end
print("\nInserting 40 and 45 at the end:")
linked_list.insert_at_end(40)
linked_list.insert_at_end(45)
linked_list.display()

# Task 3: Delete a value from the start
print("\nDeleting value from the start:")
linked_list.delete_from_start()
linked_list.display()

# Task 4: Delete a value from the end
print("\nDeleting value from the end:")
linked_list.delete_from_end()
linked_list.display()
```

```python
# You are tasked with managing a single linked list that contains a few integer values. The linked
# list allows insertion of values both at the start and the end of the list. It also supports deletion of
# values from both the start and the end of the list.
# Initially, the linked list contains the following values:
# LinkedList = [5, 10, 15, 20, 25]
# Tasks:
# 1. Insert values at the start of the linked list:
#  Insert the values 30 and 35 at the beginning of the list.
# 2. Insert values at the end of the linked list:
#  Insert the values 40 and 45 at the end of the list.
# 3. Delete a value from the start of the linked list:
#  Remove the first element from the list.
# 4. Delete a value from the end of the linked list:
#  Remove the last element from the list.
# s
# 5. Display the linked list:
#  Show the linked list before and after each update operation

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    # Function to display the linked list
    def display(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            current = self.head
            while current:
                print(current.data, end=" -> " if current.next else "\n")
                current = current.next
    
    # Function to insert at the beginning
    def insert_at_start(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
    
    # Function to insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
    
    # Function to delete from the start
    def delete_from_start(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        else:
            print(f"Deleting {self.head.data} from the start")
            self.head = self.head.next
    
    # Function to delete from the end
    def delete_from_end(self):
        if self.head is None:
            print("Linked list is empty, nothing to delete.")
        elif self.head.next is None:
            print(f"Deleting {self.head.data} from the end")
            self.head = None
        else:
            current = self.head
            while current.next.next:
                current = current.next
            print(f"Deleting {current.next.data} from the end")
            current.next = None
            
# Create a linked list and initialize it with the given values
linked_list = LinkedList()

# Inserting the initial values into the linked list
initial_values = [5, 10, 15, 20, 25]
for value in initial_values:
    linked_list.insert_at_end(value)

# Display the initial linked list
print("Initial Linked List:")
linked_list.display()

# Task 1: Insert values 30 and 35 at the start
print("\nInserting 30 and 35 at the start:")
linked_list.insert_at_start(30)
linked_list.insert_at_start(35)

# Display the linked list after inserting at the start
linked_list.display()

# Task 2: Insert values 40 and 45 at the end
print("\nInserting 40 and 45 at the end:")
linked_list.insert_at_end(40)

linked_list.insert_at_end(45)

# Display the linked list after inserting at the end
linked_list.display()

# Task 3: Delete a value from the start
print("\nDeleting value from the start:")
linked_list.delete_from_start()

# Display the linked list after deleting from the start
linked_list.display()

# Task 4: Delete a value from the end
print("\nDeleting value from the end:")
linked_list.delete_from_end()

# Display the linked list after deleting from the end
linked_list.display()

# Task 5: Display the final linked list
print("\nFinal Linked List:")
linked_list.display()
```
